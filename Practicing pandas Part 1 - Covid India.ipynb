```python
df2 = df2.fillna(0)
```

```python
df2.info()
```

```python
#df1 Assign a new variable called data1. data1 is a df that contains the rows of data whose states appear in duplicate_list.

data1 = df1.loc[df1.loc[:, 'State'].isin(duplicate_list), :]
# Next, filter data1 further so that it only contains data where the Date >= ‘2020-04-17’ and Date <= ‘2021-08-10’

data1 = data1.loc[(data1.loc[:, 'Date'] >= '2020-04-17') & (data1.loc[:, 'Date'] <= '2021-08-10'), :]
```

```python
#df2 Assign a new variable called data2. data2 is a df that contains the rows of data whose states appear in duplicate_list.
data2 = df2.loc[df2.loc[:, 'State'].isin(duplicate_list), :]
```

```python
data1 = data1.groupby('State').sum()
data1
```

```python
data2 = data2.groupby('State').sum()
data2
```

```python
#In data2, create a new column called Positive Rate, where Positive Rate = Positive / Total Samples

data2.loc[:, 'Positive Rate'] = data2.loc[:, 'Positive']/data2.loc[:, 'TotalSamples']
data2
```

```python
df_final = pd.merge(data1, data2, on ='State')
df_final
```
