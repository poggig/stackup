# Understanding the Context - Task 1
Now that you’ve loaded up your dataframe, we’re ready to take on the first task!

An inter-company Generation 1 Pokemon competition is coming up soon and your company will be participating in it. Through secret channels, your manager has learnt that the Pokemon that your company will be fighting against is Jolteon.

With this knowledge, your manager has tasked you with finding a Pokemon that is best suited for battle against it. As pokemons have different strengths and weaknesses, a good pokemon for battle against Jolteon is one where its strength is Jolteon’s weakness.

```python
import requests
from bs4 import BeautifulSoup

url='https://www.theloadout.com/pokemon-type-chart'

header = {
  "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
  "X-Requested-With": "XMLHttpRequest"
}

r = requests.get(url, headers=header)

dfs = pd.read_html(r.text)
df_weaknesses = dfs[0]
df_weaknesses.columns = df_weaknesses.iloc[0] 
df_weaknesses = df_weaknesses[1:]
```

```python
df_weaknesses.head()
```

```python
df_weaknesses[df_weaknesses['Type']=='Electric']
```

Electric is weak to Ground

```python
df_type = df.loc[df.loc[:, 'Type 1']=='Ground', :]
df_type
```

```python
df_type_gen1 = df_type.loc[df_type.loc[:, 'Generation']==1, :]
df_type_gen1
```

```python
df_type_gen1_sorted = df_type_gen1.sort_values(by='Sp. Def', ascending=False)
df_type_gen1_sorted
```

```python
sns.boxplot(x='Generation', y='HP', data=df)
```

```python
quantile_97= df.loc[:, 'HP'].quantile(0.97)
quantile_97
```

```python
hp_outliers_df = df[df['HP']>quantile_97]
hp_outliers_df.head()
```

```python
hp_outliers_id = hp_outliers_df.index.tolist() 
```

```python
df_hp_no_outliers = df.drop(hp_outliers_id, axis=0)
```

```python
df_type_gen1_sorted.iloc[[0],:]
```
